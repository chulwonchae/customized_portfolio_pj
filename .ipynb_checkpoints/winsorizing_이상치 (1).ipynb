```python
# 제외 종목 func
remove = ['141020', '103230', '058420', '263540', '160600', '114120', '050540', '058220', '009730', '121890', '180400', '223310', '138360', '038340', '003620', '069110', '058530', '208860', '257370', '056000', '102940', '175140', '033340', '299910', '099520', '006580', '033790', '141020', '058220', '121890', '160600', '263540', '109070', '223310', '058530', '257370', '003620', '056000', '058420', '101680', '006580', '103230', '065560', '178780', '103660', '033340', '078590', '069110', '099520', '038340', '012600', '015540']
def rm_list(x):
    if (x in remove):
        return False
    else:
        return True

# 종목코드를 6자리로 통일. ex) 990 -> 000990
def padding(x):
    return x.rjust(6, "0")
```

# 1. 펀더멘탈 지표 이상치 제거

```python
# 이상치 확인을 위한 boxplot 그리기 전 초기작업
import seaborn as sns #heatmap 만들기 위한 라이브러리
import matplotlib.pyplot as plt
import warnings
import numpy as np
import matplotlib
from scipy.stats.mstats import winsorize

%matplotlib inline
plt.style.use('seaborn')
warnings.simplefilter(action='ignore', category=FutureWarning)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = "NanumGothic"
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
# 파일 불러오기
fundamental = pd.read_csv('./datasets/1_fundamental.csv')
plt.figure(figsize=(20,10))
# ["NCAV", "PEGR", "PSR", "유동비율", "부채비율"]
```

```python
fundamental.head(1) # 1246개
```

```python
# [NCAV] 이상치 확인
sns.boxplot(x="NCAV", data=fundamental, palette="Set3")
sns.swarmplot(data=fundamental, x="NCAV")
```

```python
# 전체 종목 윈저라이징
# [NCAV]
fundamental['NCAV'] = winsorize(fundamental['NCAV'], limits=[0.005, 0.005])

# [PEGR]
fundamental['PEGR'] = winsorize(fundamental['PEGR'], limits=[0.005, 0.005])

# [PSR]
fundamental['PSR'] = winsorize(fundamental['PSR'], limits=[0.005, 0.005])

# [유동비율]
fundamental['유동비율'] = winsorize(fundamental['유동비율'], limits=[0.005, 0.005])

# [부채비율]
fundamental['부채비율'] = winsorize(fundamental['부채비율'], limits=[0.005, 0.005])



# 노가다 노가다
# 펀더멘탈 지표 -> 이따가 한번에 하기!!!!
fundamental = fundamental[(fundamental["NCAV"]> -2.48) & (fundamental["NCAV"] < 1.399187)] # NCAV

fundamental = fundamental[(fundamental["PEGR"]> -30) & (fundamental["PEGR"] < 20)] # PEGR

fundamental = fundamental[(fundamental["PSR"] < 60)] # PSR

fundamental = fundamental[(fundamental["유동비율"] < 1750)] # 유동비율

fundamental = fundamental[(fundamental["부채비율"] < 350)] # 부채비율


# ■최종최종최종 펀더멘탈■
fundamental
```

```python
# ★★★★★★★★ #
# [부채비율] 이상치 확인
plt.figure(figsize=(30,10))
sns.boxplot(x="부채비율", data=fundamental, palette="Set3")
sns.swarmplot(data=fundamental, x="부채비율")
```

```python
# ★★★★★★★★ #
# [유동비율] 제거대상 확인

# HEAD
# fundamental['부채비율'].sort_values(ascending=False).head(10)

# TAIL
# fundamental['유동비율'].sort_values(ascending=False).tail(10)

fundamental[(fundamental["부채비율"] < 350)]
```

```python
# 제외종목 제거하기
final_fundamental = fundamental[fundamental["종목코드"].map(rm_list)]












final_fundamental # 1197개
```

# 3. 팩터 지표 이상치 제거

```python
factor = pd.read_csv('./datasets/3_factor.csv')
# [PER PBR 시가총액]
```

```python
factor.head(1) # 2323개
```

```python
# [PER]
factor['PER'] = winsorize(factor['PER'], limits=[0.005, 0.005])

# [PBR]
factor['PBR'] = winsorize(factor['PBR'], limits=[0.005, 0.005])

# [시가총액]
factor['시가총액'] = winsorize(factor['시가총액'], limits=[0.005, 0.005])


# 노가다 노가다
# 펀더멘탈 지표 -> 이따가 한번에 하기!!!!

factor = factor[(factor["PER"] < 600)] # PER 

factor = factor[(factor["PBR"] < 70)] # PBR

factor = factor[(factor["시가총액"] <  96460313)] # 시가총액

factor
```

```python
# ★★★★★★★★ #
# [유동비율] 제거대상 확인

# HEAD
# factor['시가총액'].sort_values(ascending=False).head(10)

# TAIL
# factor['유동비율'].sort_values(ascending=False).tail(10)

# factor[(factor["시가총액"] <  96460313)]
```

```python
# 제외종목 제거하기
final_factor = factor[factor["종목코드"].map(rm_list)]












final_factor # 2310개
```

# ★. 지표 3개 합치기

```python
# final_fundamental => 1197개 
# final_danger => 1467
# final_factor =>  2310개 
```

```python
# padding 진행
final_fundamental["종목코드"] = final_fundamental["종목코드"].astype("str")
final_fundamental["종목코드"] = final_fundamental["종목코드"].map(padding)

final_danger["종목코드"] = final_danger["종목코드"].astype("str")
final_danger["종목코드"] = final_danger["종목코드"].map(padding)

final_factor["종목코드"] = final_factor["종목코드"].astype("str")
final_factor["종목코드"] = final_factor["종목코드"].map(padding)
```

```python
df_final = pd.merge(final_fundamental, final_danger, on=['종목코드'], how = 'inner')
df_final = pd.merge(df_final, final_factor, on=['종목코드'], how = 'inner')
df_final = df_final[['종목코드', '종목명', 'NCAV', 'PEGR', 'PSR', '유동비율', '부채비율', '일일변동평균', '샤프지수', 'PER', 'PBR', '시가총액']]
df_final
```

```python
# 내마음속에 저 ~ 장
df_final.to_csv('outlier_tf.csv')
```
